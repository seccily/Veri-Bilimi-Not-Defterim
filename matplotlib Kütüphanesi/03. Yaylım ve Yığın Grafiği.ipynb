# Yaylım (Nokta) ve Yığın Grafiği

## 1) Yaylım (Nokta) Grafiği

### Nokta grafikleri, nicel verileri düzeltilmemiş bir biçimde göstermek için faydalıdır.

### Nokta grafikleri, verileri bir sıra ekseni boyunca çizmek için bir çok nokta kullanır. Nokta grafiği, çizgilerin olmadığı bir çizgi grafiğiyle aynıdır. Sadece veri noktaları gösterilir.

### Aşağıdaki örnek, her bir ürün yelpazesi için geliri gösterir :

![](https://www.ibm.com/support/knowledgecenter/tr/SSEP7J_10.2.2/com.ibm.swg.ba.cognos.ug_rptstd_fin.10.2.2.doc/chart_exmpl_point.jpg)

```python
import matplotlib.pyplot as plt
```

### Önce klasik olarak kütüphanemizi import ettik. Şimdi de isimlendirmeleri yapalım ancak bu sefer farklı olarak isimlendirmelerimizi renklendirelim.

```python
x=[2,4,6,8]
y=[1,5,2,3]
```

```python
plt.title("Semih Gülüm\n matplotlib notları",color="green")
plt.xlabel("X ekseni",color="c")
plt.ylabel("Y ekseni",color="c")
plt.scatter(x,y,label="Yaylım (Nokta) Grafiği ",color='green')
plt.legend()
```

### İki farklı yaylım grafiğini tek tabloda görelim : 

```python
plt.title("Semih Gülüm\n matplotlib notları",color="green")
plt.xlabel("X ekseni",color="c")
plt.ylabel("Y ekseni",color="c")
x1=[1,2,3,4,5]
y1=[6,7,8,9,1]
y2=[3,4,5,6,7]
plt.scatter(x1,y1,label="Yaylım (Nokta) Grafiği1",color='green')
plt.scatter(x1,y2,label="Yaylım (Nokta) Grafiği2",color="red")
plt.legend()
```

### Grafiğin noktalarını veya boyutlarını değiştirmek istersek :

```python
plt.title("Semih Gülüm\n matplotlib notları",color="green")
plt.xlabel("X ekseni",color="c")
plt.ylabel("Y ekseni",color="c")
x1=[1,2,3,4,5]
y1=[6,7,8,9,1]
y2=[3,4,5,6,7]
plt.scatter(x1,y1,label="Yaylım (Nokta) Grafiği1", color="green", marker="-", s=100)
plt.scatter(x1,y2,label="Yaylım (Nokta) Grafiği2", color="red", marker="s", s=5)
plt.legend()
```

### marker cursor'ün şeklini s ise size'dan geliyor ve büyüklüğü ifade ediyor.

### NOT : 'marker' ve 'linestyles' aynı şeydir. Ben elim alıştığı için marker yazıyorum ancak siz marker yazan yerleri silip linestyles yazarak da aynı işlevde olduklarını görebilirsiniz.

> **marker="-"**              
> **linestyles="-"**

### Çıktıları aynı olacaktır.

### Marker ezberi yapmaktansa [bu linkten](https://matplotlib.org/3.1.0/api/markers_api.html?highlight=markers) hepsine bakabiliriz.

## 2) Yığın Grafiği :

### Yığın grafikleri, bir kategori içerisinde oransal katılımları karşılaştırmak için faydalıdır. Toplama eklenen her bir veri serisi göreli değeri olan göreli değeri çizer. Örneğin, ürün yelpazesi satışlarını çizen bir yığın sütun grafiği, her bölgedeki toplama eklenen her bir ürün yelpazesinin oranını vurgulayacaktır.
### Her veri serisini yığın içindeki bölüm rengi ya da kalıbı ile ayırt edebilirsiniz. Her bir yığının üst kısmı, her kategori için toplam değerleri gösterir.

![](https://www.ibm.com/support/knowledgecenter/tr/SSEP7J_10.2.2/com.ibm.swg.ba.cognos.ug_cr_rptstd.10.2.2.doc/chart_exmpl_stckd.jpg)

> Örnek bir yığın grafiği

### plt.stackplot() fonksiyonu kullanılır.

### Yeniden 24 saatimizi ele alalım.

```python
gunler=[1,2,3,4,5,6,7]   # Haftanın 7 gününü temsilen.
uyku=[7,6,8,7,7,9,10]
calisma=[10,10,10,10,10,4,0]
oyun=[7,8,6,7,7,11,14]  #amacım 24'e tamamlamak olduğu için değerler biraz mantıksız oldu.
plt.stackplot(gunler,uyku,calisma,oyun,colors=["black","red","green"])
```

### Siyah alanlar uykuyu, kırmızı alanlar calışma saatini yeşil alan ise oyunu temsil etmekte.

### 24 saate tamamladığımızı grafiğin yukarısının dümdüz olmasından anlayabiliriz. Örneğin 24 saate tamamlayamadığımız 'yanlış' bir örnek yapalım.

```python
gunler=[1,2,3,4,5,6,7]  # Haftanın 7 gününü temsilen.
uyku=[7,6,8,7,7,9,10]
calisma=[10,10,10,10,10,4,0]
oyun=[7,8,6,5,5,5,4]  # 24'e tamamlamayacak şekilde rastgele değer verdim.
plt.stackplot(gunler,uyku,calisma,oyun,colors=["black","red","green"])
```

### Grafiğin çarşambadan sonraki günlerinde 24 saati tamamlamadığımız için arka plan rengi olan beyaz renk kaldı.
