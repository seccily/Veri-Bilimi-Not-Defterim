# İlk Grafik

```python
import matplotlib.pyplot as plt
```

### matplotlib kütüphanesinin içindeki pyplot'u plt adıyla çağırmak üzere import ettik.

```python
plt.plot([3,5,7],[8,9,10])
```

### Birinci değer x'ler ikinci değer y'ler

### liste şeklinde tanımlayarak da bir atama gerçekleştirelim :

```python
xler=[1,2,3]
yler=[4,5,6]
plt.plot(xler,yler)
```

> x değerlerine karşılık gelen y değerlerinin grafiği çizilmiş oldu.

> 1 değerine 4, 2 değerine 5, 3 değerine 6 değerleri verilmiş oldu. 

## Grafiğimize başlık ve eksen isimleri koyalım : 

### x eksenine vereceğim adlandırma :

```python
plt.xlabel("x Ekseni")
```

### y eksenine vereceğim adlandırma :

```python
plt.ylabel("y Ekseni")
```

### Grafiğe başlık verme işlemini yapalım. (\n ile alt satıra geçerek başlıklandırmayı düzenleyebiliriz.)

```python
plt.title("Grafiğin Başlığı")
```

### Dikkat etmemiz gereken nokta bu başlıkları ve plt.plot()'ı aynı cell'e yazmamız gerektiğidir.

```python
plt.plot(xler,yler)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.title("Semih Gülüm \n matlpotlib Kütüphanesi")
plt.plot([3,5,7],[8,12,10])
```

### Bu iki grafiği isimlendirmek (label atamak) istersek .legen() fonksiyonuyla birlikte kullanılır. Ancak label koymazsak ve .legend()'ı kullanmak istersek bize "bana label vermedin" şeklinde bir hata mesajı basar. .legend() fonksiyonu küçük kutucuklar içerisinde bizim isimlendirmelerimizi gösterir. Bir başka deyişle grafik elemanlarını tanımlayan bileşendir de diyebiliriz. Mesela :

```python
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.title("Semih \n Gülüm")
plt.plot(xler,yler,label="Mavi Grafik")
plt.plot([3,5,7],[8,12,10],label="Kirmizi Grafik")
plt.legend()
```

### plt.legend() 'ın içine hiçbir şey yazmaksak o kendisi label'ları bulur ve kendisi otomatik olarak tabloda bir yere yerleşir. Ancak tabloda yerleşeceği yeri biz seçmek istsersek bu durumda plt.legend(loc) kullanırız. Mesela plt.legend(loc="best") yazarsak otomatik olarak kendisini en iyi yere konumlar. Örneğin :

```python
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.title("Semih \n Gülüm")
plt.plot(xler,yler,label="Mavi Grafik")
plt.plot([3,5,7],[8,12,10],label="Kirmizi Grafik")
plt.legend(loc="best")
```
