# Annotate Kavramı

### Annotate : Grafiğe yazı veya şekil ekleme işlemine verilen addır. Zaten ingilizce anlamı da *'bir metine not eklemek'*tir.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.arange(0,30,3)
```

### En küçük değeri bulalım :

```python
data.min()
```

### En büyük değeri bulalım :

```python
data.max()
```

### Önce basit bir plot grafiğini görelim :

```python
plt.plot(data)
```

### Tabloyu kendimize göre kişiselleştirelim :

```python
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(data,"k-")
x.set_title("Semih Gülüm \n matplotlib notları")
x.set_xlabel(" Girdiler ")
x.set_ylabel(" Çıktılar ")
```

### Şimdi bulduğumuz min ve max değerleri tabloda gösterelim :

```python
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(data,"k-")
x.set_title("Semih Gülüm \n matplotlib notları")
x.set_xlabel(" Girdiler ")
x.set_ylabel(" Çıktılar ")
x.annotate("Min Değer", xy=(0,0), xytext=(0,0), arrowprops=dict(facecolor="red"))
x.annotate("Max Değer", xy=(9,27), xytext=(9,27), arrowprops=dict(facecolor="c"))
```

### .annotate fonksiyonunda önce yazdırmak istediğimiz stringi, sonrasında okun bulunacağı xy koordinat sistemindeki konumunu, daha sonra da ilk argümanda girdiğimiz string ifadesinin bulunacağı konumu koordinat sisteminde belirtiriz. İsteğe bağlı olarak da *'arrowprops'* argümanı ile okun özelliklerini girebiliriz.
