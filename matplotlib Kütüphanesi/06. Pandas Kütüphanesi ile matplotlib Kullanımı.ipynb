# Pandas Kütüphanesi ile matplotlib Kullanımı

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Öncelikle basit bir çizgi grafiği çizelim. Ama ilk olarak *'Series'* tipinde bir veri oluşturalım :

```python
veri=pd.Series(np.arange(0,30,3))
veri.plot()
```

**NOT :**

x=np.arange(0,30,3)                 
veri=pd.Series(x)                                  
veri.plot()

Şeklinde uzun uzun da yazabilirdik ama artık daha az satırla daha hızlı kod yazmak için kısa yoldan yazdım.

### 4 tane rastgele sayılardan oluşan dataframe'ler oluşturalım ve bu dört dataframe'in de komülatif toplamını bularak tek tabloda A,B,C ve D şeklinde isimlendirerek görelim :

```python
dataf=pd.DataFrame(np.random.randn(100,4), columns=list("ABCD"))
dataf=dataf.cumsum()
dataf.plot()
```

### Eğer istediğimiz bir değeri belirgin bir hale getirmek istersek :

```python
dataf.iloc[10].plot(kind="bar")
```

### *'dataf'* veri setindeki bütün sütunların 10.satırını bar grafiği ile gösterdik.

### Çoklu bar grafikleri çizdirmek istersek :

```python
dataf2=pd.DataFrame(np.random.randn(7,3),columns=list("ABC"))
```

7 değer alacak şekilde 3 tane veri seti oluşturduk ve bu veri setlerinin adlarını da sırasıyla A,B ve C yaptık.

```python
dataf2.plot.bar()
```

### *'Saklanmış bar'* grafiği çizdirmek istersek :

```python
dataf2.plot.bar(stacked=True)
```

Böylelikle üst üste toplanmış bar grafikleri oluştu.

### Bar grafiğini yatayda çizdirmek istersek :

```python
dataf2.plot.barh()
```

## Histogram çizdirelim.

### Bu sefer hazır bir data üzerinden çalışacağız. İşlemlerimizi önemli bir data olan 'iris.data' üzerinden yapacağız. [İndirme linki için tıklayınız.](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/) 


### İndirmeyi tamamladıktan sonra iris.data'yı  çalışma klasörünüzün içine atınız. Şimdi işlemlerimize geçelim :

```python
iris=pd.read_csv("iris.data")
iris
```

### 149 satır çok olduğu için ben yine sadece ilk beş data ile çalışaracağım.

```python
iris.head()
```

```python
iris.plot.hist()
```

### Bu görüntü çok hoşuma gitmediği için biraz daha gölgeli bir grafik elde etmek istiyorum :

```python
iris.plot.hist(alpha=0.7)
```

### *'Saklanmış histogram'* grafiği şeklinde yapmak istersek yapacağımız işlem yine aynıdır.

```python
iris.plot.hist(alpha=0.7, stacked=True)
```

### Histogramdaki dikdörtgenlerin boyutlarını da ayarlamak istersek :

```python
iris.plot.hist(alpha=0.7, stacked=True, bins=20)
```

### Histogram grafiğini yatayda çizdirmek istersek :

```python
iris.plot.hist(alpha=0.7 ,orientation="horizontal")
```

### Veri setindeki her bir değişkenin tek seferde histogramını istersek :

```python
iris.hist(color="blue",alpha=1 ,bins=20)
```
