# Figure Nesenesi Oluşturma ve Alt Grafik Kavramı 

### Grafikler figure nesnesi ile dizayn edilirler. *plt.figure()* dediğimizde nesne oluşmuş olur.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=plt.figure()
```

### Figure ile nesene oluşturduktan sonra genellikle alt grafik denilen tabiri caizse 'grafikcik'ler eklenir.  Biz de ekleyelim :

```python
x1=a.add_subplot(2,2,1) 
```

### .add_subplot() komutu ile bahsettiğimiz grafikcikleri ekledik. Böyle dememizin nedeni normalde bir grafiklik alana küçük küçük daha çok grafik sıkıştırmamızı sağlamasından dolayıdır. Örneğin yukarıda .add_subplot(2,2,1) diyerek "Bir grafiklik alana satıra 2 sütne 2 tane gelecek şekilde 2'ye 2 kadar yani 4 tane alt grafik oluştur" demiş olduk. Sondaki 1 ise "bu 4 grafikten ilki x1'dir" dedik.

### Şimdi bir data oluşturalım. Bu data rastgele 10 sayının kümülatif toplamı olsun :

```python
data=np.random.randn(10).cumsum()
```

```python
plt.plot(data,"k*")
```

### Farklı renk ve şekillerde aynı grafiği 4 kere çizdirelim :

```python
b=plt.figure()
x1=b.add_subplot(2,2,1) 
x1.plot(data,"k*")
x2=b.add_subplot(2,2,2)
x2.plot(data,"rs")
x3=b.add_subplot(2,2,3)
x3.plot(data,"bo")
x4=b.add_subplot(2,2,4)
x4.plot(data,"c|")
```

### Şimdi ise oluşturduğumuz figure nesnesinin grafiğinin x  ekseninin değer aralıklarını kendimiz belirleyelim :

```python
data=np.random.randn(50).cumsum()
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(data)
x.set_xticks([0,15,30,45,60])
```

### Şimdi bu değerler yerine etiketler yerleştirelim.

```python
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(data)
x.set_xticklabels(["sifir","on","yirmi","otuz","kırk","elli","altmış"])
```

### Bu etiketlerin boyutunu ve yazma açısını ayarlayalım.

```python
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(data)
x.set_xticklabels(["sifir","on","yirmi","otuz","kırk","elli","altmış"], rotation=30, fontsize='small')
```

### Son dokunuş olarak da grafiğimizi kişiselleştirelim :

```python
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(data)
x.set_xticklabels(["sıfır","bir","iki","uc","dort","bes","alti"], rotation=30, fontsize='small')
x.set_title("Semih Gülüm \n matplotlib notları")
```
