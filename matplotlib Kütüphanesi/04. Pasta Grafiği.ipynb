> Örnek bir pasta grafiği

### İşin kod kısmına gelelim.

### plt.pie() fonksiyonu kullanılır.  

### Bu örneği liste oluşturup listeleri fonksiyonun içine atarak yapalım.

### Bir tv kanalında yayınlanan içeriklerin saatlerine göre dağılımını ele alalım.

```python
import matplotlib.pyplot as plt
```

```python
kategoriler=["korku","komedi","yarışma programı","dram","romantik"]
renkler=["g","m","c","r","k"]
```

### Dikkat ederseniz burada ilk defa renklere uzun uzun 'green' , 'cyan' vs. yazmak yerine tek harfle ifade ettim. Bu harf kodlarını ezberlemeye gerek yok. Ben de ihtiyacım olduğunda  [buradan](https://matplotlib.org/2.0.2/api/colors_api.html) bakıyorum.

```python
saatler=[2,4,10,5,3]  #24 saate tamamladık
```

### 'Saatler' bizim pasta grafiğimizin dilimleri olacak.

```python
plt.pie(saatler, labels=kategoriler, colors=renkler)
plt.title("Bir günlük TV yayını")
```

### Standart bir pasta grafiği yaptık.

### Eğer ilk değerin başlayacağı açıyı da belirtmek istersek startangle argümanını kullanırız.

```python
kategoriler=["korku","komedi","yarışma programı","dram","romantik"]
renkler=["g","m","c","r","k"]
saatler=[2,4,10,5,3]  ## 24 saat tamamlandı.
plt.pie(saatler, labels=kategoriler, colors=renkler, startangle=90)
plt.title("Bir Günlük TV Yayını")
```

### 'Kategoriler' değişkeninden de görülebileceği gibi benim ilk değerim "korku" kategorisi olduğu için 'korku' kategorisi 90 dereceden başlayarak pasta grafiği oluşturuldu.

### Eğer gölgeli bir pasta grafiği yapıp 3 boyutlu gibi göstermek istersek shadow argümanını kullanırız. Örneğin :

```python
kategoriler=["korku","komedi","yarışma programı","dram","romantik"]
renkler=["g","m","c","r","k"]
saatler=[2,4,10,5,3]  ## 24 saat tamamlandı.
plt.pie(saatler, labels=kategoriler, colors=renkler, shadow=True )
plt.title("Bir Günlük TV Yayını")
```

### Pasta grafiğinde seçtiğimiz bir alanı daha belirgin göstermek explode argümanı kullanırız. Örneğin :

```python
kategoriler=["korku","komedi","yarışma programı","dram","romantik"]
renkler=["g","m","c","r","k"]
saatler=[2,4,10,5,3]  ## 24 saat tamamlandı.
plt.pie(saatler, labels=kategoriler, colors=renkler, explode=(0,0,0.2,0,0))
## yarışma programını daha belirgin hale getirdim.
plt.title("Bir Günlük TV Yayını")
```

### Pasta grafiğinin değerlerini yüzdesel olarak görmek istersek autopct='%1.1f%%' argümanı kullanılır. Kalıptır.

```python
kategoriler=["korku","komedi","yarışma programı","dram","romantik"]
renkler=["g","m","c","r","k"]
saatler=[2,4,10,5,3]  ## 24 saat tamamlandı.
plt.pie(saatler, labels=kategoriler, colors=renkler, autopct='%1.1f%%')
plt.title("Bir Günlük TV Yayını")
```
