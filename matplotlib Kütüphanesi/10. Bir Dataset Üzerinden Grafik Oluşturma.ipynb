```python
data=pd.read_excel("baraj_doluluk.xlsx")
data
```

Görüldüğü üzere artık datamızı görselleştirmeye geçebiliriz.

```python
import matplotlib.pyplot as plt
```

```python
doluluk=data["Doluluk Oranı (%)"]
```

```python
plt.plot(doluluk)
```

### Tablomuza başlık bilgileri girelim ve renk,label gibi özelleştirmelerimizi yapalım :

```python
plt.title("Semih Gülüm \n matplotlib notları")
plt.xlabel(" m**3 cinsinden su miktarı ")
plt.ylabel(" Doluluk Oranı % ")
plt.plot(doluluk,label="Yıllara göre barajın doluluk oranı ",color="c")
plt.legend()
```

### x eksenindeki m**3 cinsinden su miktarını 0-5000 aralığında değil de 0-6000 aralığında göstermesini istersem .xlim fonksiyonunu kullanmam gerekir. Ancak direkt sıfırdan başlatmak yerine başta ve sonra biraz daha boşluk olmasını tercih ettiğim için x eksenini -300-6000 aralığında dizmesini söyleyeceğim.

```python
plt.title("Semih Gülüm \n matplotlib notları")
plt.xlabel(" m**3 cinsinden su miktarı ")
plt.ylabel(" Doluluk Oranı % ")
plt.xlim(-300,6000)
plt.plot(doluluk,label="Yıllara göre barajın doluluk oranı ",color="c")
plt.legend()
```

### Şimdi talodaki Tarih sütunundan yıl bilgilerini de alalım ve bunun tablosunu oluşturalım.

```python
yillar=data["Tarih"]
yillar
```

```python
plt.plot(yillar,doluluk)
```

### Yeniden tablomuzu kişiselleştirelim.

```python
plt.title("Semih Gülüm \n matplotlib notları")
plt.xlabel(" Yıllar ")
plt.ylabel(" Doluluk Oranı % ")
plt.plot(yillar,doluluk,label="Yıllara göre barajın doluluk oranı ",color="c")
plt.legend()
```

### Elimde artık iki veri olduğuna göre bar grafiği de yapalım :

```python
plt.title("Semih Gülüm \n matplotlib notları")
plt.xlabel(" Yıllar ")
plt.ylabel(" Doluluk Oranı % ")
plt.bar(yillar,doluluk,label="Yıllara göre barajın doluluk oranı ",color="c")
plt.legend()
```

### Histogram çizelim :

```python
data["Doluluk Oranı (%)"].plot.hist(alpha=0.5, bins=20, color=["green"])
```

### Yaylım (Nokta) Grafiği çizelim :

```python
plt.title("Semih Gülüm \n matplotlib notları")
plt.xlabel(" Yıllar ")
plt.ylabel(" Doluluk Oranı % ")
plt.scatter(yillar,doluluk,label="Yıllara göre barajın doluluk oranı ",color="c")
plt.legend()
```

### Yığın Grafiği Çizelim

```python
plt.title("Semih Gülüm \n matplotlib notları")
plt.xlabel(" Yıllar ")
plt.ylabel(" Doluluk Oranı % ")
plt.stackplot(yillar, doluluk, colors=["c","b"])
```

### Bu sefer de figure nesnesi oluşturarak aynı işlemleri yapalım :

```python
a=plt.figure()
x=a.add_subplot(1,1,1)
x.plot(doluluk)
```

### Bir grafiklik alana 4 tane grafik çizelim. Bu 4 grafik de doluluk oranı olsun ancak farklı şekil ve renklerde yapalım .

```python
a=plt.figure()
x1=a.add_subplot(2,2,1)
x1.plot(doluluk,"b|")
x2=a.add_subplot(2,2,2)
x2.plot(doluluk,"r--")
x3=a.add_subplot(2,2,3)
x3.plot(doluluk,"c*")
x4=a.add_subplot(2,2,4)
x4.plot(doluluk,"ks")
```
